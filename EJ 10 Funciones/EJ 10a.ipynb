```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
mu = 0
std = 1
x = np.linspace(start = -4, stop = 4, num = 100)
y = stats.norm.pdf(x, mu, std) # funcion de densidad de probabilidad
plt.plot(x,y)
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes()
plt.grid()
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes()
x = np.linspace(-5, 5, 100)
# y = np.sin(x)
plt.plot(x, np.sin(x), color = 'Indigo', linestyle = '--', linewidth = 3)
plt.grid(color='aqua', alpha = .2, linestyle='--', linewidth = 2)
plt.show()
```

```python
x = np.linspace(0,7,100)
line1, = plt.plot(x,np.sin(x),label = 'seno', color = 'orange')
line2, = plt.plot(x,np.cos(x),label = 'coseno')
plt.legend(handles=[line1, line2], loc = 'upper center')
plt.xlabel('x', fontsize = 15)
plt.ylabel('f(x)', fontsize = 15)
# grid grande
plt.grid(which = 'major', color='gray', alpha = 0.6, linestyle = 'dotted', lw = 1.5)
plt.minorticks_on()
plt.grid(which = 'minor', color='beige', alpha = 1, ls = '-', lw = 1)
plt.show()
```

```python
plt.style.use('dark_background')
mu = 0
std = 1
x = np.linspace(-4,4,100)
y = stats.norm.pdf(x, mu, std)
plt.plot(x, y, label = 'Funcion de Densidad de Probabilidad')
plt.xlabel('x', fontsize = 15)
plt.ylabel('PDF', fontsize = 15)
plt.grid(color = 'DarkTurquoise', alpha = 0.7, linestyle = ':', lw = 2)
plt.legend( loc = 'lower center')
plt.show()
```
